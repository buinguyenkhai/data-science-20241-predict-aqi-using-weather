```python
def geocode_address(city):
    # Form the API request URL
    url = f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=5&appid=bd5e378503939ddaee76f12ad7a97608'
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        # Extract relevant information
        result = data[0]
        lat = result['lat']
        lon = result['lon']
        ct_code = result['country']
        country = country = pycountry.countries.get(alpha_2 = ct_code).name
        return [city, lat, lon, ct_code, country] 
            # Return the details
    else:
        return {'error': f'Request failed with status code {response.status_code}'}
```

```python
csv_file_path = 'countries.csv'

# Open the file in write mode
with open(csv_file_path, mode='w', newline='', encoding = "utf-16") as file:
    writer = csv.writer(file)
    cities = ['London', 'Paris', 'Madrid', 'Berlin', 'Warsaw', 'Bern', 'Brussels', 'Dublin', 'Moscow', 'Rome', 'Ankara',
              'Budapest', 'Sofia', 'Amsterdam', 'Oslo', 'Stockholm', 'Helsinki', 'Copenhagen', 'Tallinn', 'Vienna', 'Zagreb',
              'Lisbon', 'Kyiv', 'Athens', 'Prague', 'Tirana', 'Bucharest', 'Belgrade', 'Podgorica', 'Nicosia', 'Ljubljana', 'Luxembourg',
                'Reykjavik', 'Bratislava', 'Minsk','Sarajevo', 'Vilnius', 'Chișinău', 'Riga', 'Skopje', 
                'Vaduz', 'Tbilisi', 'Yerevan', 'Baku', 'Monaco', 'San Marino', 'Tórshavn', 'Andorra la Vella', 'Gibraltar']
    writer.writerow(['city','lat','lon','ct_code', 'country'])
    for city in cities:
        writer.writerow(map(str, geocode_address(city)))
    
```
