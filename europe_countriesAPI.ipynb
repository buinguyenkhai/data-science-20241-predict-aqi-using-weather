```python
def geocode_address(city):
    small_caps = ['Vatican City', 'Longyearbyen', 'Mariehamn', 'Saint Helier', 'Douglas', 'St Peter Port']
    if city in small_caps:
        return
    # Form the API request URL
    url = f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=5&appid=bd5e378503939ddaee76f12ad7a97608'
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        # Extract relevant information
        result = data[0]
        lat = "%.4f"% result['lat']
        lon = "%.4f"% result['lon']
        ct_code = result['country']
        country = country = pycountry.countries.get(alpha_2 = ct_code).name
        return [city, lat, lon, ct_code, country] 
            # Return the details
    else:
        return {'error': f'Request failed with status code {response.status_code}'}
```

```python
username = 'LucasLocker222'
url = f'http://api.geonames.org/countryInfoJSON?username={username}'

response = requests.get(url)
data = response.json()

countries = []
capitals = []
continents = []
if 'geonames' in data:
    transcontinental_cnt = ['Türkiye', 'Georgia', 'Armenia', 'Azerbaijan']
    for item in data['geonames']:
        country_name = item.get('countryName')
        capital_name = item.get('capital')
        continent_name = item.get('continent')
        if country_name and capital_name and continent_name:
            countries.append(country_name)
            capitals.append(capital_name)
            continents.append(continent_name)

df = pd.DataFrame({'city': capitals, 'country': countries, 'continent': continents})
df['country'] = df['country'].str.strip()
small_countries = ['Åland', 'Guernsey', 'Jersey', 'Svalbard and Jan Mayen', 'Holy See (Vatican City State)', 'Isle of Man']
final_df = df[(df['continent'] == 'EU') | 
               (df['country'].isin(transcontinental_cnt)) & 
               (~df['country'].isin(small_countries))]
final_df.to_csv('countries.csv', index = False)
```

```python
csv_file_path = 'european_countries.csv'

# Open the file in write mode
with open(csv_file_path, mode='w', newline='', encoding = "utf-16") as file:
    writer = csv.writer(file)
    cities = final_df['city']
    writer.writerow(['city','lat','lon','ct_code', 'country'])
    for city in cities:
        try:
            writer.writerow(map(str, geocode_address(city)))
        except Exception as e:
            print() 
```
