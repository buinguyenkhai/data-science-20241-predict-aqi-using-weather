```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/countries/worldcities.csv")
capltals = df1[df1["capital"] == "primary"]
capltals
```

```python
df2 = pd.read_csv("data/countries/countries.csv")
europe_country = df2[df2["continentName"] == "Europe"]["isoAlpha3"]
europe_country
```

```python
europe_capitals = capltals[capltals["iso3"].isin(europe_country.values)]
europe_capitals
```

```python
europe_capitals.to_csv("data/region/europe/cities.csv", index=False)
```
